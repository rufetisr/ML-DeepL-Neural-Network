```python
import numpy as np

def gradient_descent(x, y):
  m_curr = b_curr = 0  # m - k - coefficient (weight), b - intercept (bias), y = k*x + b
  iterations = 3000
  learning_rate = 0.01 # when learning rate increases we close faster to minimum loss value

  prev_cost = 0.0 # initial value prev_cost

  for i in range(iterations):
    y_predicted = m_curr * x + b_curr
    md = -(2/len(x)) * sum(x * (y - y_predicted)) # m derivative
    bd = -(2/len(x)) * sum(y - y_predicted) # b derivative
    m_curr = m_curr - learning_rate * md # adjust the weight
    b_curr = b_curr - learning_rate * bd # adjust the bias
    # cost function or loss function is same thing. Mean squared error is one of loss function.
    cost = (1/len(x)) * sum([val**2 for val in (y-y_predicted)])

    print("m {}, b {}, cost {}, iteration {}".format(m_curr, b_curr, cost, i))

    if np.isclose(cost, prev_cost, rtol=1e-06): # if prev_cost and cost value difference so little it applies early stopping
      print(f'early stopping at iteration: {i}')
      break
    prev_cost = cost

x = np.array([1,2,3,4,5]) # inputs
y = np.array([5,7,9,11,13]) # output

gradient_descent(x,y)

```
